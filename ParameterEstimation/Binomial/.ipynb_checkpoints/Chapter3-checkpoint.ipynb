## 3.2 - Difference between two rates

```python
spec_string = """
model:
#Rate_2.txt
model{
   # Observed Counts
   k1 ~ dbin(theta1,n1)
   k2 ~ dbin(theta2,n2)
   # Prior on Rates
   theta1 ~ dbeta(1,1)
   theta2 ~ dbeta(1,1)
   # Difference Between Rates
   delta <- theta1-theta2
}


settings:
chains = 2
samples = 10000
thinning = 0
burnin = 0

data:
k1 = 14
n1 = 20

k2 = 16
n2 = 20

sample:
theta1
theta2
delta
"""

model_spec = ModelHandler(spec_string)

pyjags_model = pyjags.Model(**model_spec.get_model_args())
sample_data = pyjags_model.sample(**model_spec.get_sample_args())

sample_handler = SampleHandler(sample_data)
```

```python


sample_handler.vizchains('theta1','theta2', range=105)
sample_handler.vizhist('theta1')
sample_handler.vizkde('delta', showmax=True, range=(-1,1))
```

```python
# Custom plots, not built into the SampleHandler.
# For these, we'll need to import additional plotting
# modules. 

import seaborn as sns
import matplotlib.pyplot as plt

fig, ax1 = plt.subplots()

pstpk = sample_handler.samples['postpredk']
prepk = sample_handler.samples['priorpredk']

sns.countplot(pstpk, color='red', alpha=.5, ax=ax1);
sns.countplot(prepk, color='blue', alpha=.5, ax=ax1);

fig, ax1 = plt.subplots()
sns.distplot(sample_handler.get('theta'), color='red', ax=ax1);
sns.distplot(sample_handler.get('thetaprior'), color='blue', ax=ax1);
```

## 3.6 - Joint distributions

```python
import numpy as np
```

```python
spec_string = """
model:
Survey.txt

settings:
chains = 4
samples = 10000
thinning = 0
burnin = 0

data:
nmax = 500
m = 5
k = [16, 18, 22, 25, 27]

sample:
n
theta


"""

model_spec = ModelHandler(spec_string)

pyjags_model = pyjags.Model(**model_spec.get_model_args())
sample_data = pyjags_model.sample(**model_spec.get_sample_args())

sample_handler = SampleHandler(sample_data)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
#I'm likley to move this code into pjbcmasssistant.sample_handler
#if it comes up again.

from scipy.special import gammaln

samplebuddy = sample_handler
modelbuddy = model_spec

samplebuddy.vizchains('theta',range=100)


#+++++++++++++++++++++++++++++++++++
# Maximum Likelihood adapted from matlab script survey.m
# I emphatically do NOT know why this produces the answer it does.
# ---what is the gamma function doing here?
# ---this method appears to break if there's too little data?
# ---does this keep coming up? Should it be put in ModelHandler?
cc = -np.Inf 
ind = 0 

n = samplebuddy.get('n')
k = modelbuddy.data['k']
theta = samplebuddy.get('theta')
nsamples = len(samplebuddy.get('theta'))
m = int(modelbuddy.data['m'])

for i in range(nsamples): 
    logL=0 
    for j in range(m): 

        logL=logL+gammaln(n[i]+1)-gammaln(k[j]+1)-gammaln(n[i]-k[j]+1)
        logL=logL+k[j]*np.log(theta[i])+(n[i]-k[j])*np.log(1-theta[i])

        
    if logL > cc:
        ind=i
        cc=logL
#+++++++++++++++++++++++++++++++++++

#plot the joint distribution
joint = sns.jointplot('n','theta',data=samplebuddy.samples, color="grey");


#plot the Mode, as our mysteriously derived MLE
joint.ax_joint.plot(samplebuddy.get('n')[ind], samplebuddy.get('theta')[ind], 'go--', linewidth=5, markeredgewidth=2, markersize=7, marker="o", markeredgecolor='green',markerfacecolor='white');

#plot naive expectation as mean(n) by mean(theta)
joint.ax_joint.plot(samplebuddy.get('n').mean(), samplebuddy.get('theta').mean(), 'go--', linewidth=5, markeredgewidth=2, markersize=12, marker="x", markeredgecolor='red');


print(f"MLE at n={samplebuddy.get('n')[ind]}, rate={samplebuddy.get('theta')[ind]}")
```
