## 3.4 - Prior and posterior prediction

```python
settings = """
model:
Rate_4.txt

settings:
chains = 2
samples = 10000
thinning = 0
burnin = 0

data:
k = 1
n = 20

sample:
theta
thetaprior
postpredk
priorpredk

"""

modelbuddy = model_handler(settings)

#print(modelbuddy.model)

model = pyjags.Model(**modelbuddy.definition())
samples = model.sample(**modelbuddy.sample_rules())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

## 3.6 - Joint distributions

```python
import numpy as np
```

```python
settings = """
model:
Survey.txt

settings:
chains = 4
samples = 10000
thinning = 0
burnin = 0

data:
nmax = 500
m = 5
k = [16, 18, 22, 25, 27]

sample:
n
theta


"""

modelbuddy = model_handler(settings)

#modelbuddy.data['k']=obs

#print(modelbuddy.model)

model = pyjags.Model(**modelbuddy.definition())
samples = model.sample(**modelbuddy.sample_rules())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#I'm likley to move this code into pjbcmasssistant.sample_handler
#if it comes up again.

from scipy.special import gammaln

samplebuddy = sample_handler(samples)
samplebuddy.vizchains('theta',range=100)


#+++++++++++++++++++++++++++++++++++
# Maximum Likelihood adapted from matlab script survey.m
# I emphatically do NOT know why this produces the answer it does.
# ---what is the gamma function doing here?
# ---this method appears to break if there's too little data?
# ---does this keep coming up? Should it be put in modelbuddy?
cc = -np.Inf 
ind = 0 

n = samplebuddy.get('n')
k = modelbuddy.data['k']
theta = samplebuddy.get('theta')
nsamples = len(samplebuddy.get('theta'))
m = int(modelbuddy.data['m'])

for i in range(nsamples): 
    logL=0 
    for j in range(m): 

        logL=logL+gammaln(n[i]+1)-gammaln(k[j]+1)-gammaln(n[i]-k[j]+1)
        logL=logL+k[j]*np.log(theta[i])+(n[i]-k[j])*np.log(1-theta[i])

        
    if logL > cc:
        ind=i
        cc=logL
#+++++++++++++++++++++++++++++++++++

#plot the joint distribution
joint = sns.jointplot('n','theta',data=samplebuddy.samples, color="grey");


#plot the Mode, as our mysteriously derived MLE
joint.ax_joint.plot(samplebuddy.get('n')[ind], samplebuddy.get('theta')[ind], 'go--', linewidth=5, markeredgewidth=2, markersize=7, marker="o", markeredgecolor='green',markerfacecolor='white');

#plot naive expectation as mean(n) by mean(theta)
joint.ax_joint.plot(samplebuddy.get('n').mean(), samplebuddy.get('theta').mean(), 'go--', linewidth=5, markeredgewidth=2, markersize=12, marker="x", markeredgecolor='red');


print(f"MLE at n={samplebuddy.get('n')[ind]}, rate={samplebuddy.get('theta')[ind]}")
```
