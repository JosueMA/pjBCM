# Chapter 3 - Parameter Estimation

```python
import pyjags
from pjbcmassistant import model_spec
from pjbcmassistant import visualizer
```

## 3.1 - Inferring a rate

```python
model_handler = model_spec(model="Rate_1.txt")

nchains = 2
nsamples = 10000

data = dict(
k = 0,
n = 1,
)

data2 = 
"""
model = Rate_1.txt

settings:
chains = 1
samples = 1000
thinning = none
burnin = none

data:
k = 0
n = 1

sample:
theta
"""

model = pyjags.Model(code=model_handler.model, data=data, chains=nchains, init=None)
samples = model.sample(nsamples, vars=['theta'])
```

```python
viz = visualizer(samples)

viz.chains('theta', range=105)
viz.hist('theta')
viz.kde('theta', showmax=True)
```

## 3.2 - Difference between two rates

```python
model_handler = model_spec(model="Rate_2.txt")

nchains = 3
nsamples = 50000

data = dict(
k1 = 5,
k2 = 7,

n1 = 10,
n2 = 10,
)

#DRY: don't put data variables in data dict manually. do this with the helper

model = pyjags.Model(code=model_handler.model, data=data, chains=nchains, init=None)
samples = model.sample(nsamples, vars=['theta1','theta2'])
```
